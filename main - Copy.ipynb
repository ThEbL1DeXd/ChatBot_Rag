```python
import os
from dotenv import load_dotenv
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("cv.pdf")
pages = loader.load()

print(pages[0].page_content)
```

```python
import re
import unicodedata

def clean_text(text):
    """Normalize text by removing noisy chars and extra whitespace."""
    if text is None:
        return ""

    text = str(text)

    # Normalize unicode forms (quotes, spaces, compatibility chars).
    text = unicodedata.normalize("NFKC", text)

    # Unify line breaks, then flatten to single spaces.
    text = text.replace("\r\n", "\n").replace("\r", "\n").replace("\n", " ")

    # Remove common zero-width/invisible characters.
    text = re.sub(r"[\u200B-\u200D\u2060\uFEFF]", "", text)

    # Remove control characters and replacement char from broken encodings.
    text = re.sub(r"[\x00-\x1F\x7F\uFFFD]", " ", text)

    # Collapse repeated whitespace and trim.
    text = re.sub(r"\s+", " ", text).strip()

    return text
cleaned_pages = [clean_text(page.page_content) for page in pages]
print(cleaned_pages[0])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
# Utilisation de l'optimisation Tiktoken
token_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
encoding_name="cl100k_base",
chunk_size=500,
chunk_overlap=50
)
rec_docs = token_splitter.split_documents(pages)
print("Number of documents after splitting :",len(rec_docs))
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
print("Chargement du modèle local (BAAI/bge-m3)...")
model_local = SentenceTransformer('BAAI/bge-m3')
```

```python
texts=[doc.page_content for doc in rec_docs]
embeddings_local = model_local.encode(texts)
print(f"Dimension d'un vecteur BAAI/bge-m3 : {len(embeddings_local[0])}")
```

```python
import chromadb
import uuid

print("Initialisation de la base de donnees vectorielle...")

db_path = "./ma_base_vectorielle"

# 1. Initialiser le client (les donnees seront sauvegardees sur le disque)
client = chromadb.PersistentClient(path=db_path)

# 2. Creer ou recuperer une collection (equivalent d'une table SQL)
collection = client.get_or_create_collection(
    name="index_bge_m3",
    metadata={"hnsw:space": "cosine"}
 )

# 3. Générer des identifiants uniques pour chaque chunk
ids = [str(uuid.uuid4()) for _ in range(len(texts))]

# 4. Insérer les données dans la collection
print(f"Indexation de {len(texts)} documents en cours...")
collection.add(
    documents=texts,     
    embeddings=embeddings_local, 
    ids=ids                
)
print(f" Succès ! Les données sont sauvegardées dans le dossier '{db_path}'.")
print(f"Collection '{collection.name}' contient maintenant {collection.count()} elements.")
```

```python
# --- 1. Paramètres de la recherche ---
question = "linkedin link"
k = 3  # Le nombre de textes pertinents qu'on veut récupérer

print(f"Recherche des {k} passages les plus pertinents pour : '{question}'")

# --- 2. Transformer la question en vecteur ---
question_embedding = model_local.encode([question])[0]

# --- 3. Interroger la base vectorielle ---
results = collection.query(
    query_embeddings=[question_embedding.tolist()],
    n_results=k
)

# --- 4. Extraire et afficher les résultats ---
relevant_chunks = results['documents'][0]
distances = results['distances'][0]
print(results)
print("\n✅ Recherche terminée !")
for i, chunk in enumerate(relevant_chunks):
    print(f"\n--- Résultat {i+1} (Distance: {distances[i]:.4f}) ---")
    print(chunk[:150] + "...") 
```

```python
# --- 5. Préparation du contexte ---
context = "\n\n".join(relevant_chunks)

# --- 6. Construction du Prompt ---
prompt = f"""Tu es un assistant expert et précis. Ton rôle est de répondre à la question de l'utilisateur en te basant UNIQUEMENT sur le contexte fourni ci-dessous.
Règles à respecter impérativement :
1. N'utilise que les informations présentes dans la section CONTEXTE.
2. Si la réponse ne se trouve pas dans le contexte, dis simplement \"Je ne trouve pas la réponse dans les documents fournis\". N'invente jamais d'informations (pas d'hallucination).
3. Sois clair, concis et structure ta réponse.

CONTEXTE :
{relevant_chunks}

QUESTION : 
{question}

RÉPONSE :
"""

print(" Prompt construit avec succès !\n")
```

```python
from groq import Groq

load_dotenv()

API_KEY = os.getenv("Groq_API_Key")

client = Groq(api_key=API_KEY)

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "user", "content": prompt}
    ],
    temperature=0.1
)

print(response.choices[0].message.content)
```
